# Using Support vector Machine

```python


import seaborn as sns  
import pandas as pd  
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions


```

```python


dataset=pd.read_csv("placement-dataset.csv")
sns.scatterplot(data=dataset,x="cgpa",y="iq",hue="placement")
plt.show()
```

```python


x=dataset.iloc[:,1:-1]
y=dataset["placement"]


```

```python
x
```

```python


from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)


```

```python


from sklearn.svm import SVC
s=SVC(kernel="linear")
s.fit(x_train,y_train)
s.score(x_test,y_test)
```

```python
s.score(x_train,y_train)
```

```python
''' if you see any kind of overfitting in model then
try changing random_state value or try changing kernel
'''
```

```python

plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=s)
plt.show()
```

# using logistic regression

```python
sns.scatterplot(data=dataset,x="cgpa",y="iq",hue="placement")
plt.legend(loc=1)
'''Displays a legend that explains the
color coding for the hue variable (placed).loc=1
places the legend in the top-right corner of the
plot. (Other values for loc specify different
positions.)
'''
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
lr.score(x_test,y_test)
```

```python
lr.coef_

```

```python
lr.intercept_
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=lr)# asks for x,y,clf and all data should be in array
plt.show()
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
gnb=GaussianNB()
gnb.fit(x_train,y_train)
gnb.score(x_test,y_test)
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=gnb)
plt.show()
```

```python
# trying multinomial
mnb=MultinomialNB()
mnb.fit(x_train,y_train)
mnb.score(x_test,y_test)
```

```python
plt.figure(figsize=(4,5))
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=mnb)
plt.show()
```

```python
# trying bernoulli
bnb=BernoulliNB()
bnb.fit(x_train,y_train)
bnb.score(x_test,y_test)
```
